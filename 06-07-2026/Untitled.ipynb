```python
#image colour Chainging !!!
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
img = Image.open("colour.jpg")
img = np.array(img)
h, w, _ = img.shape
mid_h = h // 2
mid_w = w // 2
img[:mid_h, :mid_w, 0] = np.clip(img[:mid_h, :mid_w, 0] + 100, 0, 255)
img[:mid_h, mid_w:, 1] = np.clip(img[:mid_h, mid_w:, 1] + 100, 0, 255)
img[mid_h:, :mid_w, 2] = np.clip(img[mid_h:, :mid_w, 2] + 100, 0, 255)
img[mid_h:, mid_w:, 0] = np.clip(img[mid_h:, mid_w:, 0] + 255, 255, 255)
plt.imshow(img)
plt.axis("off")
plt.show()
```
