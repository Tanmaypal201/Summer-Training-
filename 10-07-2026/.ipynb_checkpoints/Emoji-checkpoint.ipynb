```python
clc;
clear;
close all;

figure;
hold on;
axis equal;
grid on;

% House Body (Square)
plot([2 8],[2 2],'k','LineWidth',2);   % Bottom
plot([2 2],[2 8],'k','LineWidth',2);   % Left
plot([8 8],[2 8],'k','LineWidth',2);   % Right
plot([2 8],[8 8],'k','LineWidth',2);   % Top

% Roof
plot([2 5],[8 11],'r','LineWidth',2);
plot([5 8],[11 8],'r','LineWidth',2);

% Door
plot([4 4],[2 5],'b','LineWidth',2);
plot([6 6],[2 5],'b','LineWidth',2);
plot([4 6],[5 5],'b','LineWidth',2);

% Left Window
plot([2.8 4.2],[5 5],'g','LineWidth',2);
plot([2.8 2.8],[5 6.5],'g','LineWidth',2);
plot([4.2 4.2],[5 6.5],'g','LineWidth',2);
plot([2.8 4.2],[6.5 6.5],'g','LineWidth',2);

% Right Window
plot([5.8 7.2],[5 5],'g','LineWidth',2);
plot([5.8 5.8],[5 6.5],'g','LineWidth',2);
plot([7.2 7.2],[5 6.5],'g','LineWidth',2);
plot([5.8 7.2],[6.5 6.5],'g','LineWidth',2);

% Door Knob (Circle)
rectangle('Position',[5.75 3.2 0.15 0.15], ...
          'Curvature',[1 1], ...
          'FaceColor','k');

title('Simple House using Coordinates, Lines and Circle');
hold off;
```
