```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle, Polygon, Circle

fig, ax = plt.subplots(figsize=(7, 7))
ax.add_patch(Rectangle((0, 0), 10, 10, color='#AEDFF7', zorder=0))
ax.add_patch(Rectangle((0, 0), 10, 2, color='#8FBF6D', zorder=0))
wall = Rectangle((2, 2), 6, 4, facecolor='#F4E1B7', edgecolor='black', linewidth=2, zorder=1)
ax.add_patch(wall)
roof = Polygon([[1.5, 6], [8.5, 6], [5, 9]], closed=True,
               facecolor='#B5432A', edgecolor='black', linewidth=2, zorder=2)
ax.add_patch(roof)
door = Rectangle((4.3, 2), 1.4, 2.2, facecolor='#5C3A21', edgecolor='black', linewidth=2, zorder=3)
ax.add_patch(door)
ax.add_patch(Circle((5.5, 3.1), 0.06, color='gold', zorder=4))
for x in [2.7, 6.3]:
    win = Rectangle((x, 4), 1.2, 1.2, facecolor='#A6D8F0', edgecolor='black', linewidth=2, zorder=3)
    ax.add_patch(win)
    ax.plot([x + 0.6, x + 0.6], [4, 5.2], 'k', lw=1.5, zorder=4)
    ax.plot([x, x + 1.2], [4.6, 4.6], 'k', lw=1.5, zorder=4)
chimney = Rectangle((6.5, 7.2), 0.6, 1.3, facecolor='#8B4A3A', edgecolor='black', linewidth=1.5, zorder=2)
ax.add_patch(chimney)
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.set_aspect('equal')
ax.axis('off')
plt.title("Normal House", fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Wedge, Ellipse

fig, axes = plt.subplots(1, 2, figsize=(10, 5))

def add_gradient_sphere(ax, base_color):
    ax.add_patch(Circle((0, 0), 1.0, facecolor=base_color, edgecolor='#00000030', linewidth=1, zorder=1))
    ax.add_patch(Ellipse((-0.35, 0.45), 0.45, 0.28, angle=25, facecolor='white', alpha=0.35, zorder=2))
ax = axes[0]
add_gradient_sphere(ax, '#FFCC4D')
ax.add_patch(Ellipse((-0.32, 0.22), 0.16, 0.20, facecolor='#3A2C1A', zorder=3))
ax.add_patch(Ellipse((0.32, 0.22), 0.16, 0.20, facecolor='#3A2C1A', zorder=3))
ax.add_patch(Ellipse((-0.55, -0.15), 0.22, 0.13, facecolor='#FF9AA2', alpha=0.6, zorder=2))
ax.add_patch(Ellipse((0.55, -0.15), 0.22, 0.13, facecolor='#FF9AA2', alpha=0.6, zorder=2))
smile = Wedge((0, -0.05), 0.55, 195, 345, width=0.28, facecolor='#3A2C1A', zorder=3)
ax.add_patch(smile)
ax.add_patch(Wedge((0, -0.05), 0.27, 200, 340, width=0.05, facecolor='#C1392B', alpha=0.5, zorder=4))
ax.set_xlim(-1.3, 1.3); ax.set_ylim(-1.3, 1.3)
ax.set_aspect('equal'); ax.axis('off')
ax.set_title("😄 Happy", fontsize=16, fontweight='bold')
ax = axes[1]
add_gradient_sphere(ax, '#F4900C')
ax.add_patch(Ellipse((-0.32, 0.15), 0.14, 0.14, facecolor='#3A2C1A', zorder=3))
ax.add_patch(Ellipse((0.32, 0.15), 0.14, 0.14, facecolor='#3A2C1A', zorder=3))
ax.plot([-0.58, -0.15], [0.52, 0.28], color='#3A2C1A', lw=9, solid_capstyle='round', zorder=3)
ax.plot([0.58, 0.15], [0.52, 0.28], color='#3A2C1A', lw=9, solid_capstyle='round', zorder=3)
ax.plot([-0.06, -0.02], [0.35, 0.22], color='#3A2C1A', lw=2, zorder=3)
ax.plot([0.06, 0.02], [0.35, 0.22], color='#3A2C1A', lw=2, zorder=3)
frown = Wedge((0, -0.62), 0.42, 25, 155, width=0.16, facecolor='#3A2C1A', zorder=3)
ax.add_patch(frown)
ax.set_xlim(-1.3, 1.3); ax.set_ylim(-1.3, 1.3)
ax.set_aspect('equal'); ax.axis('off')
ax.set_title("😠 Angry", fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```
