```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
```

```python
df = pd.read_csv("HeartDisease.csv")
```

```python
df.head()
```

```python
print(df["Smoking"].unique())
print(df["HeartDisease"].unique())
```

```python
df["Smoking"] = df["Smoking"].map({"Yes": 1, "No": 0})
df["AlcoholDrinking"] = df["AlcoholDrinking"].map({"Yes": 1, "No": 0})
df["Stroke"] = df["Stroke"].map({"Yes": 1, "No": 0})
df["Diabetic"] = df["Diabetic"].map({"Yes": 1, "No": 0})
df["DiffWalking"] = df["DiffWalking"].map({"Yes": 1, "No": 0})
df["PhysicalActivity"] = df["PhysicalActivity"].map({"Yes": 1, "No": 0})
df["HeartDisease"] = df["HeartDisease"].map({"Yes": 1, "No": 0})
```

```python
df.isnull().sum()#Checking NULL
df.head()
df["HeartDisease"].unique()
```

```python
df.dropna(inplace = True) #Exact match
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
x = df[[
    "BMI",
    "Smoking",
    "Stroke",
    "Diabetic",
    "PhysicalActivity",
    "DiffWalking"
]]

y = df["HeartDisease"]
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , Y_train, Y_test = train_test_split(x , y , test_size = 0.2 , random_state = 42)
#X_TEST=[set of 20 % of the X ] 
#X_TRAIN=[set of 80% of the X]
#Y_TEST=[set of 20% of the Y ] OUTPUT !! 
#Y_TRAIN=[set of 80% of the Y] Output WE train it !! 
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train , Y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score , classification_report
```

```python
accuracy = accuracy_score(y_test, y_pred)
```
